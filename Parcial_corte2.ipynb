```python
import pandas as pd

productos = pd.read_csv("parcial_2_productos_sucios.csv")
clientes = pd.read_csv("parcial_2_clientes_sucios.csv")
proveedores = pd.read_csv("parcial_2_proveedores_sucios.csv")

print(productos.head())
print(clientes.head())
print(proveedores.head())
```

```python
productos[['producto', 'categoria']] = productos['producto_categoria'].str.split(' - ', expand=True)

productos.drop(columns=['producto_categoria'], inplace=True)

productos['precio'] = productos['precio'].replace('[\$,]', '', regex=True).astype(float)

productos['stock'] = productos['stock'].fillna(0)

productos['producto'] = productos['producto'].str.title()
productos['categoria'] = productos['categoria'].str.title()

print(productos.head())
```

```python
clientes[['nombre', 'tipo']] = clientes['cliente_tipo'].str.split(' - ', expand=True)

clientes.drop(columns=['cliente_tipo'], inplace=True)

clientes.drop(columns=['edad'], inplace=True)

clientes['telefono'] = clientes['telefono'].fillna("No Registra")
clientes['email'] = clientes['email'].fillna("No Registra")

clientes['nombre'] = clientes['nombre'].str.title()
clientes['tipo'] = clientes['tipo'].str.title()

print(clientes.head())
```

```python
proveedores[['empresa', 'ciudad']] = proveedores['empresa_ciudad'].str.split(' - ', expand=True)

proveedores.drop(columns=['empresa_ciudad'], inplace=True)

proveedores['telefono'] = proveedores['telefono'].fillna("No Registra")
proveedores['email'] = proveedores['email'].fillna("No Registra")
proveedores['contacto'] = proveedores['contacto'].fillna("No Registra")

proveedores['empresa'] = proveedores['empresa'].str.title()
proveedores['ciudad'] = proveedores['ciudad'].str.title()

print(proveedores.head())
```

```python
productos.to_csv("parcial_2_productos_limpios.csv", index=False)
clientes.to_csv("parcial_2_clientes_limpios.csv", index=False)
proveedores.to_csv("parcial_2_proveedores_limpios.csv", index=False)
```
